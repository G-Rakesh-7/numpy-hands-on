# 2. Create numpy array that have zero, one, two,three dimentions respectivily. Also, print the shape and dimentions of the numpy arrays created above.

```python
import numpy as np
# 1-D
a1 = np.array([1, 2, 3, 4, 5])
print('1D Array:',a1)
print('Shape:', a1.shape)
print('Dimensions:', a1.ndim)
# 2-D
a2= np.array([[1, 2, 3], [4, 5, 6]])
print('\n2D Array:',a2)
print('Shape:', a2.shape)
print('Dimensions:', a2.ndim)
# 3-D
a3 = np.array([[[1, 2],[3, 4]],[[5, 6],[7, 8]]])
print('\n3D Array:',a3)
print('Shape:', a3.shape)

print('Dimensions:', a3.ndim)
```

```python
a1d = np.random.randint(1,10,5)

a2d = np.random.randint(1,10,(4,3))

a3d = np.random.randint(1,10,(5,2,2)) # here (5,2,2) is (Depth , row , column)
print(a1d,a2d,a3d)

print(a1d.shape,a2d.shape,a3d.shape)
print(a1d.ndim,a2d.ndim,a3d.ndim)
```
